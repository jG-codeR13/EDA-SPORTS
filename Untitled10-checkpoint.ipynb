```python
import math
def computeTicks (x, step = 5):
    """
    Computes domain with given step encompassing series x
    @ params
    x    - Required - A list-like object of integers or floats
    step - Optional - Tick frequency
    """
    xMax, xMin = math.ceil(max(x)), math.floor(min(x))
    dMax, dMin = xMax + abs((xMax % step) - step) + (step if (xMax % step != 0) else 0), xMin - abs((xMin % step))
    return range(dMin, dMax, step)
```

```python
batsmen_summary = deli.groupby("batsman").agg({'ball': 'count','batsman_runs': 'sum'})
batsmen_summary.rename(columns={'ball':'balls', 'batsman_runs': 'runs'}, inplace=True)
batsmen_summary = batsmen_summary.sort_values(['balls','runs'], ascending=False)
batsmen_summary['batting_strike_rate'] = batsmen_summary['runs']/batsmen_summary['balls'] * 100
batsmen_summary['batting_strike_rate'] = batsmen_summary['batting_strike_rate'].round(2)
batsmen_summary.head(10)
```

```python
batsmen_summary['runs'].dtype
```

```python
N = 3
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)
def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                ha='center', va='bottom')


for i in batsmen_summary['batsman']:
    yvals = batsmen_summary['runs'][i]
    rects1 = ax.bar(ind, yvals, width, color='r')
    zvals = batsmen_summary['balls'][i]
    rects2 = ax.bar(ind+width, zvals, width, color='g')
    kvals = batsmen_summary['runs'][i]
    rects3 = ax.bar(ind+width*2, kvals, width, color='b')

    ax.set_ylabel(batsmen_summary['batting_strike_rate'])
    ax.set_xticks(ind+width)
    ax.set_xticklabels( batsmen_summary['batsman'] )
    ax.legend( (rects1[0], rects2[0], rects3[0]), ('y', 'z', 'k') )


    autolabel(rects1)
    autolabel(rects2)
    autolabel(rects3)

plt.show()
```
